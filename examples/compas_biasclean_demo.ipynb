## Overview

This notebook demonstrates the complete BiasClean v3.0 audit-first fairness pipeline on the COMPAS recidivism dataset.

## The Four Execution Modes (Source: Chapter 33.3, Table 33.1, p. 297)

| Mode | Governance Gate | Mitigation | SVM Enforcement | Output |
|------|----------------|------------|-----------------|--------|
| **Audit-Only** | ✅ Yes | ❌ No | ❌ No | Audit log only |
| **Audit-First** | ✅ Yes | ✅ Yes (conditional) | ❌ No | Corrected dataset + report |
| **Legacy (No SVM)** | ❌ No | ✅ Yes | ❌ No | Corrected dataset + report |
| **Legacy (+ SVM)** | ❌ No | ✅ Yes | ✅ Yes | Corrected dataset + model outputs |

## Step 4: Mode 2 – Audit-First (Governance Gate + Conditional Mitigation)

**Source:** Chapter 33.3, p. 296–297

> *"Audit-First also begins with a pre-mitigation audit and generates the same traffic-light governance assessment. However, the pipeline does not terminate after issuing the governance signal. Instead, the governance outcome is recorded as a formal checkpoint, and the system proceeds to data-level mitigation."*

**Use case:** Embed governance oversight before intervention while still allowing remediation.

```python
# Initialize pipeline in audit-first mode (RECOMMENDED)
pipeline_audit_first = UniversalBiasClean(
    domain='justice',
    mode='audit_first',
    enable_svm=False
)

# Process dataset (replace df with actual data)
# results_audit_first = pipeline_audit_first.process_dataset(
#     df=df,
#     target_column='two_year_recid'
# )

# Expected results from Chapter 36, p. 326
print("=" * 60)
print("MODE 2: AUDIT-FIRST (RECOMMENDED)")
print("=" * 60)

# Governance gate output
print("--- GOVERNANCE GATE ---")
print("Traffic Light: YELLOW")
print("Action: REVIEW REQUIRED BEFORE MITIGATION")
print("Vulnerable groups identified: Ethnicity, Age, Gender")

# After governance approval (assuming mitigation proceeds)
print("\n--- AFTER GOVERNANCE APPROVAL ---")
print("Proceeding to data-level mitigation...")

print("\n--- MITIGATION OUTCOMES (Source: Chapter 36, p. 326) ---")
print(f"Initial Bias Score: 0.2213")
print(f"Final Bias Score: 0.1591")
print(f"Bias Reduction: 28.1%")
print(f"Data Retention: 101.4% (controlled synthesis applied)")

print("\n--- FEATURE-LEVEL IMPROVEMENTS ---")
print("Ethnicity: 30.1% improvement")
print("Age: 30.4% improvement")
print("Gender: 2.5% improvement")

print("\nArtifacts generated:")
print("- corrected_dataset.csv")
print("- biasclean_report.html")
print("- pipeline_summary.json")
```
